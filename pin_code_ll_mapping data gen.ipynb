```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page

import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

import folium # map rendering library

print('Libraries imported.')
```

```python
url = 'https://indiamapia.com/Mumbai.html'

results = requests.get(url).text
```

```python
soup = BeautifulSoup(results,"html5lib")
```

```python
extract_data = soup.find_all("tbody")
```

```python
extract_data[1]
```

```python
table_data = extract_data[1].find_all("tr")
```

```python
table_data[0]
```

```python
link = table_data[0].find("a").get("href")
print(link)
```

```python
results2 = requests.get('https://indiamapia.com'+link).text
soup2 = BeautifulSoup(results2,"html5lib")
data2 = soup2.find_all("b")
```

```python
print(data2[10],data2[12])
```

```python
pin_code_data = pd.DataFrame(columns = ['AreaName','PinCode','Lat','Long'])

for row in table_data:
    
    data = row.find_all("a")
    
    area = data[0].text.strip()
    pincode = data[1].text.strip()
    
    print("Area: {}, Pincode: {}".format(area, pincode))
    
    link = row.find("a").get("href")
    
    url2 = 'https://indiamapia.com' + link
    results2 = requests.get(url2).text
    soup2 = BeautifulSoup(results2,"html5lib")
    data2 = soup2.find_all("b")
    
    try:
        lat = data2[10].text.strip()
        long = data2[12].text.strip()
        print("Lat: {}, Long: {}".format(lat, long))
        pin_code_data = pin_code_data.append({'AreaName':area,'PinCode':pincode,'Lat':lat,'Long':long}, ignore_index=True)
    except:
        print("Oops! Error point skipped")
        print(data2)
    
pin_code_data.head()
    
```

```python
pin_code_data.to_csv('pin_code_ll_mapping.csv')
```
